```python
import numpy as np
import matplotlib.pyplot as plt
import json
import pydde as d
```

```python
#Parameters
samplenum = 1
epochs = 200
input_size = 3
output_size = 3
learning_rate = 0.01
nTimeSteps = 3; #seconds
```

```python
dyn = d.PyDyn('../Data/rb-pendulum/twoRb.sim', nTimeSteps)
```

```python
# load parameters
with open('../Data/rb-pendulum/twoRb.p') as f:
	data = json.load(f)
p = np.array(data['parameters'])
```

```python
# compute trajectory
state = dyn.compute(p)
```

```python
# plot x trajectory of 2nd rigid body
plt.plot(state.y[6::12])
plt.show()
```

```python
# plot x velocity of 2nd rigid body
plt.plot(state.ydot[6::12])
plt.show()
```
